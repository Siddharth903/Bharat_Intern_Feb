```python
data_2= data[['bedrooms','bathrooms','floors','sqft_living','condition','street','price']]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=4600 , stop_words='english')
cv
```

```python
vector = cv.fit_transform(data_2['price'].values.astype('U')).toarray()
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cs = cosine_similarity(vector)
cs
```

```python
data_2[data_2['condition']==4].index[0]
```

```python
dist = sorted(list(enumerate(cs[2])), reverse=True , key=lambda vector:vector[1])
for i in dist[0:1]:
    print(data_2.iloc[i[0]].condition)
```

```python
def recommend(data):
    index=data_2[data_2['condition']==data].index[0]
    dist = sorted(list(enumerate(cs[index])),reverse=True , key = lambda vector: vector[1])
    for i in dist[0:1]:
        print(data_2.iloc[i[0]].price)
```

```python
recommend(4)
```

```python
recommend(1)
```

```python
recommend(3.0)
```
